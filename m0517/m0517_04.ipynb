```python
from sklearn import svm,metrics
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import pandas as pd
import random

matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 10
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_csv('../z20_data/02.iris.csv')
df.head(3)
```

```python
df.shape
```

```python
df.iloc[:,4:]
```

```python
# 1. 데이터 전처리
# 데이터 분리 SepalLength,SepalWidth,PetalLength,PetalWidth/ Name
datas = df.iloc[:,:-1]
results = df.iloc[:,4:]
```

```python
# train, test 데이터 분리 - 자동으로 랜덤으로 섞여 분리해줌
train_input,test_input,train_target, test_target = train_test_split(datas,results)
```

```python
test_input
```

```python
# 2.데이터 학습시키기
clf = svm.SVC()
clf.fit(train_input, train_target)

# 3. 데이터 예측하기
pre = clf.predict(test_input)
print("test질문의 답 : ",pre)
```

```python
len(test_input)
```

```python
# 정답률 구하기
score = clf.score(test_input,test_target)
print("정답률 : ",score)
```
