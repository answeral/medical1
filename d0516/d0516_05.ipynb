```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import pandas as pd
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 10
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_excel('../z20_data/score.xlsx',index_col='지원번호')
df['학년'] = [3,3,2,1,1,3,2,2]
df.head()
```

```python
plt.bar(df['이름'],df['국어'])
plt.plot(df['이름'],df['영어'])
```

```python
# subplots 여러개 그래프 생성
fig,axs = plt.subplots(2,2,figsize = (10,10))

# 0,0번째 그래프 작성
# 범례 : label
axs[0,0].bar(df['이름'],df['국어'],label='국어점수')
# 범례출력
axs[0,0].legend()
# 상단타이들
axs[0,0].set_title('국어점수그래프')
# x축, y축 글자표시
axs[0,0].set(xlabel='이름',ylabel='국어')
# 그래프 배경색상
axs[0,0].set_facecolor('#efefef')
# 격자표시
axs[0,0].grid(linestyle='--',alpha=0.5)

# 0,1번째 그래프 작성
axs[0,1].plot(df['이름'],df['영어'],label='영어점수')
axs[0,1].plot(df['이름'],df['수학'],label='수학점수')
axs[0,1].set_title('영어, 수학점수그래프')
axs[0,1].legend(loc='upper center')

# 1,0 그래프 작성
axs[1,0].barh(df['이름'],df['사회'])
axs[0,1].set_title('사회점수그래프')
# 1,1 그래프 작성
axs[1,1].plot(df['이름'],df['과학'],color='green',alpha=0.4)
axs[0,1].set_title('과학점수그래프')

plt.show()
```
