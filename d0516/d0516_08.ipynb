```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import pandas as pd
import seaborn as sns
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 10
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
#데이터 불러오기
df = pd.read_csv('../z20_data/diabetes.csv')
df
```

```python
# 나이가 50이상인 사람
(df['Age']>=50).value_counts()
```

```python
# 나이가 50이상이면서 당뇨병 환자인 사람
df[df[df['Age']>=50] & df['Outcome'] ==1]
df[df['Age']>=50 & df['Outcome'] ==1].count()
```

```python
# 각 항목 별 상관관계 함수 :1과 가까울 수록 상관관께가 있다고 판단
df.corr()
          
```

```python
data_corr = df.corr()
fig = plt.figure(figsize=(16,16))
sns.heatmap(data_corr,annot=True) # 각 항목별 데이터를 출력시켜줌.
plt.show()
```

```python
### outcome??을 가지고 원 그래프를 출력

# df[df['Outcome']==1].count()
values = df['Outcome'].value_counts().values
labels = ['비환자','환자']

```

```python
explode=[0,0.1]
plt.title('당뇨병 환자분포')
plt.pie(values,labels=labels,autopct='%.1f%%',explode=explode,startangle=90)
plt.show()
```
