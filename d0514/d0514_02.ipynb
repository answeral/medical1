```python
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 12
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
x = ['han','jim','hong']
y = [20,40,60]
plt.plot(x,y)

plt.title('선그래프',fontdict={'family':'HYGungSo-Bold','size':30})
# x,y축 설명문구, 위치, 색상
plt.xlabel('이름',loc='right',color='red') # left, center, right
plt.ylabel('국어',loc='top',color = 'green') # top, center, bottom
```

```python
x = ['han','jim','hong']
y = [20,40,60]
y2 = [50,30,80]
y3 = [60,20,90]

plt.plot(x,y2)
plt.plot(x,y3)

# 범례 이름 지정
plt.plot(x,y,label='국어점수')
plt.plot(x,y2,label='영어점수')
plt.plot(x,y3,label='수학점수')
# 범례 표시, 위치설정, ncol 1줄에 범례를 n개씩 출력
# plt.legend(loc='upper left')
plt.legend(loc=1,ncol=3)

plt.title('선그래프')
# x,y축 설명문구, 위치, 색상
plt.xlabel('이름',loc='right',color='red') # left, center, right
plt.ylabel('국어',loc='top',color = 'green') # top, center, bottom
```
