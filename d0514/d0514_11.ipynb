```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import pandas as pd
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 10
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
labels = [30,25,20,13,10,2] #100
# labels = [1,1,1,1,1,1] #100
title = ['python','java','javascript','C#','C','ETC']
# explode = [0.1]*6
explode = [0,1,0,0,0,0]
# startangle : 시작위치 
# 데이터표시 : autopct
# counterclock=False:  90도를 기준으로 시계방향
# 파이 사이 간격 : explode
plt.pie(labels,labels=title,autopct='%.1f%%',explode=explode,startangle=90,counterclock=False)
plt.title('전체 언어별 선호도')
plt.legend(loc=(1.2,0.3),title='언어별 선호도')
plt.show()
```
