```python
# 전체 평균보다 맥주 소비량이 높은 대륙을 출력하시오.
beer_mean = df['beer_servings'].mean()
avg_beer = df.groupby('continent')['beer_servings'].mean() # 전체 평균
high = avg_beer[avg_beer >= beer_mean]

# 전체 평균보다 맥주소비량이 높은 대륙 검색
high
```
