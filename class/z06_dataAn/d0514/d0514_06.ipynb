```python
# 합계 컬럼을 생성해서 막대그래프로 출력
import pandas as pd
df = pd.read_excel('../z20_data/score.xlsx',index_col='지원번호')
df['합계'] = df['과학']+df['국어']+df['사회']+df['수학']+df['영어']
df
```

```python
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 10
matplotlib.rcParams['axes.unicode_minus'] = False
plt.figure(dpi=150)
x = df['이름']
y = df['합계']
plt.barh(x,y,label='합계')
plt.legend(loc='upper right')
plt.title('합계')

for i , txt in enumerate(y):
     plt.text(y[i]+13.0,x[i],txt,ha='center',color='green')
     
plt.xticks([100,200,300,400,500])
# plt.xlim(90,500)
plt.show()
```
