```python
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 10
matplotlib.rcParams['axes.unicode_minus'] = False

x = [1,2,3]
y = [2,4,8]
```

```python
plt.plot(x,y,label='국어',linestyle='-.')
plt.legend(loc='upper right')
plt.show()
```

```python
# 화면 해상도 크기
# plt.figure(dpi=200)
plt.plot(x,y,label='국어',linestyle='-.',marker='o')
plt.legend(loc='upper right')
# 파일지정
# plt.savefig('g2.png',dpi=200)

# 좌표에 데이터가 2,4,8의값을 출력
# plt.text()
# x = [1,2,3] y = [2,4,8] : 데이터값
plt.xticks([0,1,2,3,4,5])
plt.yticks([x for x in range(0,11)])
for i, txt in enumerate(y):
     plt.text(x[i],y[i]+0.2,txt,ha='center',color='blue')
plt.show()
```

```python
import pandas as pd
df = pd.read_excel('../z20_data/score.xlsx',index_col='지원번호')
df
```

```python
# 키를 꺾은 선 그래프로 출력
# 이름, 키
# y축 표시 170,180,190,200,210
plt.figure(dpi=150)
x = df['이름']
y = df['키']
plt.plot(x,y,label='키',linestyle='-.',marker='o',mfc='yellow',linewidth='5', ms=10)
plt.legend(loc='upper center')
# plt.yticks([150,160,170,180,190,200,210]) # 그래프 y축 눈금 표시
for i, txt in enumerate(y):
     plt.text(x[i],y[i]+2.0,txt,ha='center',color='blue')

plt.ylim(165,210) # 눈금 범위 지정
plt.show()
```

```python
# 키를 막대그래프로 출력
# 이름, 키
# y축 표시 170,180,190,200,210
plt.figure(dpi=150)
x = df['이름']
y = df['키']
# barh 옆으로 막대 그래프
plt.barh(x,y,label='키')
plt.legend(loc='upper center')
for i, txt in enumerate(y):
     plt.text(y[i]+1.5,x[i],txt,ha='center',color='blue')

plt.xticks([150,160,170,180,190,200,210]) # 그래프 y축 눈금 표시
plt.xlim(165,210) # 눈금 범위 지정
plt.show()
```
