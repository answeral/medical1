```python
# 막대그래프 score.xlsx 합계 점수를 뽑아서 그래프로 표현
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import pandas as pd
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 10
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
import pandas as pd
df = pd.read_excel('../z20_data/score.xlsx')
df['합계'] = df['과학']+df['국어']+df['사회']+df['영어']+df['수학']
df
```

```python
x = df['이름']
y = df['합계']

plt.bar(x,y,label='합계')
plt.legend(loc='upper center')
plt.title('점수 합계')
for i, txt in enumerate(y):
     plt.text(x[i],y[i]+4.0,txt,ha='center',color='green')

plt.ylim(50,500)
#  그래프 안내선 출력
plt.grid(axis='y',alpha=1.0,linestyle='--',color='black')
plt.show()
```
