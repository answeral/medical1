```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import pandas as pd
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 10
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
import pandas as pd
df = pd.read_csv('../z20_data/stu_score.csv')
df.head(10)
```

```python
h_df = df.head(20)
x = df['id'].head(20)
y = df['total'].head(20)
y1 = df['kor'].head(20)
y2 = df['eng'].head(20)
y3 = df['math'].head(20)
```

```python
plt.bar(x,y)
plt.ylim(120,300)
plt.xticks(rotation=50)
plt.show()
```

### 다중 막대 그래프 

```python
h_df.shape[0]
```

```python
idx = np.arange(h_df.shape[0])
idx
```

```python
plt.figure(figsize=(12,5))
plt.bar(idx-0.4,y1,width=0.3) #국어
plt.bar(idx,y2,width=0.3) #영어
plt.bar(idx+0.4,y3,width=0.3) #수학
for i , txt in enumerate(y2):
     plt.text(idx[i],y2[i],txt,ha='center')
plt.xticks(idx,x,rotation=50)

plt.show()
```
