```python
import matplotlib.pyplot as plt
# 한글설정
import matplotlib
matplotlib.rcParams['axes.unicode_minus'] = False # 마이너스 표시
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # window 사용자
matplotlib.rcParams['font.size'] = '13' # 글자크기

x = [1,2,3]
y = [2,4,8]
y2 = [1,4,9]
y3 = [12,15,19]

# 그래프 그리기
plt.plot(x,y)
plt.plot(x,y2)
plt.plot(x,y3)
plt.title('선 그래프')
plt.show() # 그래프만 화면에 보임. 실행하는 문구가 사라짐.
```
