```python
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 10
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
import pandas as pd
df = pd.read_excel('../z20_data/score.xlsx',index_col='지원번호')
df['합계'] = df['과학']+df['국어']+df['사회']+df['수학']+df['영어']
```

```python
x = df.head(8)['이름']
y = df.head(8)['국어']
y1 = df.head(8)['영어']
y2 = df.head(8)['수학']
```

```python
plt.plot(x,y,label='국어',marker='D')
plt.plot(x,y1,label='영어',marker='o',linestyle='--')
plt.plot(x,y2,label='수학',marker='s',ls='-.')
# ncol 범례 1줄에 3개까지 가능
for i , txt in enumerate(y2):
     plt.text(x[i],y2[i]+1.0,txt, ha ='center')
plt.legend(loc='upper center',ncol=3)
plt.show()

```

```python
color = ['r','g','b','y','olive']
plt.bar(x,y,label='국어',color=color)
# plt.plot(x,y1,label='영어',marker='o',linestyle='--')
# plt.plot(x,y2,label='수학',marker='s',ls='-.')

# ncol 범례 1줄에 3개까지 가능
for i , txt in enumerate(y2):
     plt.text(x[i],y2[i]+1.0,txt, ha ='center')
plt.legend(loc='upper right',ncol=3)
plt.xticks(rotation=45)
plt.show()


```
