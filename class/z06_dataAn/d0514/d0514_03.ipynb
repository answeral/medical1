```python
import pandas as pd
df = pd.read_excel('../z20_data/score.xlsx',index_col='지원번호')
df['학년'] = [3,3,2,1,1,3,2,2]
df
```

```python
# 이름, 국어성적 선그래프 그리기
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] =15
matplotlib.rcParams['axes.unicode_minus'] = False

x = df['이름']
y = df['국어']
plt.bar(x,y,label='국어점수')
# plt.plot(x,y,label='국어점수') 꺾은 선 그래프
plt.title('성적그래프')
plt.legend(loc='upper center')
```

```python
#chipotle.tsv에서 주문 중 가장 많이 주문한 메뉴 10개를 꺾은선 그래프로 출력하시오
import pandas as pd
df = pd.read_csv('../z20_data/chipotle.tsv',sep='\t')
df
```

```python
df.groupby('item_name').sum().count()
df.groupby('item_name')['quantity'].sum().sort_values(ascending=False).head()

```

```python
df['item_name'].value_counts()[:5]
```

```python
df.groupby('item_name')['quantity'].sum().head(10).sort_values(ascending=False)
x = df.groupby('item_name')['quantity'].sum().head(5).sort_values(ascending=False).index
y = df.groupby('item_name')['quantity'].sum().head(5).sort_values(ascending=False).values
y2 = df['item_name'].value_counts()[:5]
df.groupby('item_name').sum()
plt.bar(x,y,label='주문수')
plt.plot(x,y2,label='주문횟수')
plt.legend(loc='upper right')
```
