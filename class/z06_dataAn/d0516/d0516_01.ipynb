```python
# 막대그래프 score.xlsx 합계 점수를 뽑아서 그래프로 표현
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import pandas as pd
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 10
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
import pandas as pd
df = pd.read_excel('../z20_data/score.xlsx',index_col='지원번호')
df['학년'] = [3,3,2,1,1,3,2,2]
df.head()
```

```python
plt.bar(df['이름'],df['국어'])
```

```python
plt.plot(df['이름'],df['영어'])
plt.plot(df['이름'],df['수학'])
```

```python
#산점도
plt.scatter(df['영어'],df['수학'])
plt.xlabel('영어점수')
plt.ylabel('수학점수')
```

```python
plt.scatter(df['영어'],df['수학'])
plt.scatter(df['영어'],df['과학'])
```

```python
sizes = np.random.rand(8)*1000 #0-1사이에 랜덤 8개 수가 나옴
plt.scatter(df['영어'],df['수학'],s= 500) # s : 산저모 그래프의 점의 크기 조정
```

```python
# 학년별 크기 조절  1*500 , 2*500 3*500
sizes = df['학년']*500 
plt.scatter(df['영어'],df['수학'],s= sizes) 
plt.xlabel('영어점수')
plt.ylabel('수학점수')
```

```python
# 학년별 크기 조절  1*500 , 2*500 3*500
sizes = df['학년']*500 
plt.scatter(df['영어'],df['수학'],s= sizes,c=df['학년'],cmap='viridis') 
plt.xlabel('영어점수')
plt.ylabel('수학점수')
# 컬러바 눈금표시 :ticks, 라벨표시 : label, 컬러바크기조절 : shrink , 컬러바 위치 :orientation
plt.colorbar(ticks=[1,2,3],label='학년',shrink=0.5,orientation='horizontal')
```
