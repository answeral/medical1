```python
import pandas as pd
df = pd.read_excel('../z20_data/score.xlsx',index_col='지원번호')
df
```

### 조건 검색부분

```python
# 2개 이상 조건 검색
lang =['Python','java']
filt = df['SW특기'].isin(lang)
df[filt]
```

```python
# startswith , contains, lower, upper
# 문자열 함수적용 - lower() : 모두 소문자로 바꿔줌
# 문자열 함수적용 - upper() : 모두 대문자로 바꿔줌
lang =['python','java']
df['SW특기'].str.lower()
df['SW특기'].str.upper()

# 2개이상의 검색을 할 수 있도록 함. isin 명령어 사용 
filt = df['SW특기'].str.lower().isin(lang)
df[filt]
```

```python
# java라는 글자가 포함되어 있는 것을 검색
# na= False, Nan데이터를 True, False로 처리해줌
# filt1 = df['SW특기'].str.contains('java',na=False)
filt1 = df['SW특기'].str.lower().str.contains('java',na=False)
filt1
df[filt1]
```

```python
import pandas as pd
df = pd.read_csv('../z20_data/stu_score.csv',index_col='no')
df

## 이름에 'i'가 들어가 있는 학생검색
i_id = df['id'].str.lower().str.contains('i')
df[i_id]

## 평균 67점이상인 학생
i_avg = df['avg'] >= 80
df[i_avg]

## 둘 다 
df[i_id & i_avg]

#### 이름에 all 이 포함되어 있으면서 평균이 67점 이상인 학생을 검색하시오

a_id = df['id'].str.lower().str.contains('all')

len(df[a_id & i_avg])
df[a_id & i_avg]
```
